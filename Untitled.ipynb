```python
import torch
from torch.utils.data import TensorDataset, DataLoader

from src.nomto import NOMTO
from src.nomto_config import NOMTOConfig
from src.nomto_training import NOMTOTrainingConfig, train_nomto


nomto_cfg = NOMTOConfig()

X_train = torch.randn(1024, nomto_cfg.n_input_fields, 256)
y_train = torch.randn(1024, nomto_cfg.n_output_fields, 256)

train_dataset = TensorDataset(X_train, y_train)

train_loader = DataLoader(
    train_dataset,
    batch_size=32,
    shuffle=True,
)

model = NOMTO(nomto_cfg)

train_cfg = NOMTOTrainingConfig(
    device=nomto_cfg.device,
    phase1_epochs=100,
    phase2_epochs=100,
    phase3_epochs=100,
    l1_reg_coeff_phase2=1e-5,
    prune_every_epochs_phase2=20,
    pruning_fraction_phase2=0.1,
    pruning_min_active_edges=20,
    print_every=10,
)

loss_fn = torch.nn.MSELoss()

model, history = train_nomto(
    model=model,
    dataloader=train_loader,
    loss_fn=loss_fn,
    cfg=train_cfg,
)
```
