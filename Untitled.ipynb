```python
from __future__ import annotations

from dataclasses import dataclass, field
from pathlib import Path

import numpy as np
import sympy as sp
import torch
from torch.utils.data import DataLoader, TensorDataset

from src.nomto import NOMTO
from src.nomto_training import NOMTOTrainingConfig, train_nomto


@dataclass
class FractionalDiffusionNOMTOPathsConfig:
    simulation_data_dir: Path = Path("simulations/data")
    dataset_filename: str = "fractional_diffusion_dataset.npz"

    neural_operator_dir: Path = Path("neural_operator_surrogates")
    neural_operator_models_dir: Path = Path("neural_operator_surrogates/models")

    output_dir: Path = Path("nomto_results/fractional_diffusion")


@dataclass
class FractionalDiffusionNOMTOConfig:
    n_input_fields: int = 1
    n_output_fields: int = 1
    n_symbolic_layers: int = 1

    weight_scale: float = 0.5
    device: str = "cuda:0"

    batch_size: int = 2**14
    train_fraction: float = 1.0

    paths: FractionalDiffusionNOMTOPathsConfig = field(
        default_factory=FractionalDiffusionNOMTOPathsConfig
    )

    operation_specs: list[list[dict]] = field(
        default_factory=lambda: [
            [
                # {"type": "exact", "name": "mul"},
                # {"type": "exact", "name": "square"},
                # {
                #     "type": "neural",
                #     "name": "Dx",
                #     "task_name": "first_derivative",
                #     "model_name": "fno",
                #     "arity": 1,
                # },
                # {
                #     "type": "neural",
                #     "name": "Dxx",
                #     "task_name": "second_derivative",
                #     "model_name": "fno",
                #     "arity": 1,
                # },
                {
                    "type": "neural",
                    "name": "FracLap",
                    "task_name": "fractional_laplacian",
                    "model_name": "fno",
                    "arity": 1,
                },
                # {
                #     "type": "neural",
                #     "name": "PoissonGrad",
                #     "task_name": "poisson_gradient",
                #     "model_name": "fno",
                #     "arity": 1,
                # },
            ],
            # [
            #     {"type": "exact", "name": "mul"},
            #     {"type": "exact", "name": "square"},
            #     {
            #         "type": "neural",
            #         "name": "Dx",
            #         "task_name": "first_derivative",
            #         "model_name": "fnn",
            #         "arity": 1,
            #     },
            #     {
            #         "type": "neural",
            #         "name": "Dxx",
            #         "task_name": "second_derivative",
            #         "model_name": "fnn",
            #         "arity": 1,
            #     },
            #     {
            #         "type": "neural",
            #         "name": "FracLap",
            #         "task_name": "fractional_laplacian",
            #         "model_name": "fnn",
            #         "arity": 1,
            #     },
            #     {
            #         "type": "neural",
            #         "name": "PoissonGrad",
            #         "task_name": "poisson_gradient",
            #         "model_name": "fnn",
            #         "arity": 1,
            #     },
            # ],
        ]
    )


def load_fractional_diffusion_dataset(cfg: FractionalDiffusionNOMTOConfig):
    dataset_path = cfg.paths.simulation_data_dir / cfg.paths.dataset_filename
    data = np.load(dataset_path)

    u = data["u"].astype(np.float32)
    t = data["t"].astype(np.float32)

    dt = float(t[1] - t[0])

    u_mid = u[:, 1:-1, :]
    du_dt = (u[:, 2:, :] - u[:, :-2, :]) / (2.0 * dt)

    X = u_mid.reshape(-1, u.shape[-1])
    y = du_dt.reshape(-1, u.shape[-1])

    X = X[:, None, :]
    y = y[:, None, :]

    X = torch.tensor(X, dtype=torch.float32)
    y = torch.tensor(y, dtype=torch.float32)

    return X, y


def make_dataloader(
    X: torch.Tensor,
    y: torch.Tensor,
    cfg: FractionalDiffusionNOMTOConfig,
):
    n_samples = X.shape[0]
    n_train = int(cfg.train_fraction * n_samples)

    X = X[:n_train]
    y = y[:n_train]

    dataset = TensorDataset(X, y)

    return DataLoader(
        dataset,
        batch_size=cfg.batch_size,
        shuffle=True,
    )


def save_results(
    model: NOMTO,
    history: dict[str, list[float]],
    expression,
    cfg: FractionalDiffusionNOMTOConfig,
):
    cfg.paths.output_dir.mkdir(parents=True, exist_ok=True)

    checkpoint_path = cfg.paths.output_dir / "nomto_fractional_diffusion.pt"
    history_path = cfg.paths.output_dir / "training_history.npz"
    expression_path = cfg.paths.output_dir / "symbolic_expression.txt"

    torch.save(
        {
            "model_state_dict": model.state_dict(),
            "config": cfg,
            "expression": str(expression),
        },
        checkpoint_path,
    )

    np.savez_compressed(
        history_path,
        total_loss=np.array(history["total_loss"]),
        data_loss=np.array(history["data_loss"]),
        l1_loss=np.array(history["l1_loss"]),
        active_edges=np.array(history["active_edges"]),
        phase=np.array(history["phase"]),
    )

    with open(expression_path, "w") as f:
        f.write(str(expression))
        f.write("\n")

    return checkpoint_path, history_path, expression_path


def main():
    nomto_cfg = FractionalDiffusionNOMTOConfig()

    X_train, y_train = load_fractional_diffusion_dataset(nomto_cfg)

    train_loader = make_dataloader(
        X=X_train,
        y=y_train,
        cfg=nomto_cfg,
    )

    model = NOMTO(nomto_cfg)

    train_cfg = NOMTOTrainingConfig(
        device=nomto_cfg.device,
        seed=42,
        learning_rate=1.0e-2,
        weight_decay=0.0,
        phase1_epochs=1000,
        phase2_epochs=1000,
        phase3_epochs=1000,
        l1_reg_coeff_phase2=1e-3,
        prune_every_epochs_phase2=100,
        pruning_fraction_phase2=0.3,
        pruning_min_active_edges=5,
        print_every=100,
    )

    loss_fn = torch.nn.MSELoss()

    model, history = train_nomto(
        model=model,
        dataloader=train_loader,
        loss_fn=loss_fn,
        cfg=train_cfg,
    )

    model.prune_by_threshold(1.0e-3)

    u = sp.Symbol("u")
    expression = model.get_symbolic_output(
        symbolic_inputs=[u],
        rounding_decimals=4,
    )[0]

    checkpoint_path, history_path, expression_path = save_results(
        model=model,
        history=history,
        expression=expression,
        cfg=nomto_cfg,
    )

    print("Finished NOMTO fractional diffusion experiment.")
    print(f"Training samples: {X_train.shape[0]}")
    print(f"Input shape: {tuple(X_train.shape)}")
    print(f"Target shape: {tuple(y_train.shape)}")
    print(f"Active edges: {model.count_active_edges()}")
    print(f"Discovered expression for u_t:")
    print(expression)
    print(f"Saved checkpoint to: {checkpoint_path}")
    print(f"Saved history to: {history_path}")
    print(f"Saved expression to: {expression_path}")


if __name__ == "__main__":
    main()
    
```

```python
expression_path = Path("nomto_results/fractional_diffusion/symbolic_expression.txt")

expression_string = expression_path.read_text().strip()

u = sp.Symbol("u")

local_dict = {
    "u": u,
    "Dx": sp.Function("Dx"),
    "Dxx": sp.Function("Dxx"),
    "FracLap": sp.Function("FracLap"),
    "PoissonGrad": sp.Function("PoissonGrad"),
}

expression = sp.sympify(expression_string, locals=local_dict)

print("Discovered SymPy expression for u_t:")
sp.pprint(expression)

print("\nRaw SymPy expression:")
print(expression)

print("\nLaTeX expression:")
print(sp.latex(expression))
```
